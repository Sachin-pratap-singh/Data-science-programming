    
3. Build a SVM Classifier, train the model
   

```python
from sklearn.svm import LinearSVC
clf_svc=LinearSVC(C=5,max_iter=2000)
```

```python
#from sklearn.svm import SVC
#clf_svc=SVC(kernel="linear", degree=1, coef0=1,  C=50)
clf_svc=clf_svc.fit(X_train,y_train)
```

```python
clf_svc
```

 3.a Evaluate your model based on applicable metrics. Show the metric(s) you chose and why you chose this(these) metrics.
   

```python
from sklearn.model_selection import cross_val_predict
y_pred = cross_val_predict(clf_svc, X_test, y_test, cv=3)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

### The Evaluation metrics that i am choosing here is mean_absolute_error,mean_Squared_error, recall and precision. as mean squared error will suggest us how good we are in terma of distance between acrual and predicted.
### Precision and recall will help us understand the tweaking we need to do in threshold to control the precision or recall on the case by case basis.
### In the current scenario, we want to get an alarm for a building which is not safe or weak or vulnerable for earthquake, Here we do not want to predict weak building as strong strictly but we may accept strong building to predict as weak. Because labeling a weak building then taking actions may prevent damage.
### In other words we want low false positives or High precision.(strong building=positive, weak building=negative)
### So,To understand the model's performance and applicability we need to keep these metrics.

```python
from sklearn.metrics import precision_score, recall_score, mean_absolute_error, mean_squared_error
print("Precision Score : ",precision_score(y_test, y_pred, pos_label='positive',average='micro'))
print("Recall Score : ",recall_score(y_test, y_pred, pos_label='positive',average='micro'))
print("mean_absolute_error",mean_absolute_error(y_test,y_pred))
print("mean_squared_error",mean_squared_error(y_test,y_pred))
```

3.b List the hyper-parameters that can be tuned in SVM Classifier. Explain the meaning of each hyper-parameter. Show the improvement you achieved in model accuracy.


Ans:
Hyperparamaeter that we may tune to get impacts in SVM Classifier primarily are: 

penalty(l1 or l2): To do Regularization. Basically to penalize the cost function, and address the overfitting.
 
loss:(hinge or squared hinge)Specifies the loss function.
 
dual:(bool, default=True) algorithm to solve dual or primal optimization problem. 
tol: stopping criteria
C: Regularization parameter, more the value of c lower the strength of regularization. 
multi_class:(ovr, crammer_singer) determines the multiclass strategy when y has more than two classes, crammer_singer is rarely used.optimizes a joint objective over all classes.

fit_intercept: whether to use intercept or not 

intercept_scaling: To lessen the effect of regularization on synthetic feature weight (and therefore on the intercept)  intercept_scaling has to be increased.
 
class_weight: To set the class weight.
 
verbose:enable verbose output
 
random_state: controls the random number generation for shuffling the data if dual = True, basically it generates the same indexes everytime.
 
max_iter= maximum number of iteration to be run.

### In the model i tried afew combination of parameter values and got the best results with c=5, max_iter=2000



 3.c Plot Learning curve on the optimized model and provide insights

```python
#Plot Learning curve on the optimized model and provide insights
from sklearn.model_selection import learning_curve
import warnings
warnings.filterwarnings('ignore')
```

```python
train_siz, train_scores, validation_scores = learning_curve(estimator=clf_svc, X=X_train, y=y_train, cv=2, train_sizes=np.linspace(0.05,1,40),scoring="neg_mean_squared_error")
```

```python
train_scores_mean = -train_scores.mean(axis = 1)
validation_scores_mean = -validation_scores.mean(axis = 1)
plt.plot(train_scores_mean, label = 'Training error')
plt.plot(validation_scores_mean, label = 'Validation error')
plt.title("Learning Curve")
plt.xlabel("Training Set")
plt.ylabel("Error")
plt.legend(loc="lower right")
```


### Learning curve suggests that training and validation error is almost close to each other in thew right side of the plot which suggests that this model doesn't show significant overfitting.
